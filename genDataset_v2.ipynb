```python
total_time = 50
total_users = 3
cpw = np.zeros([total_time+1, total_users])
mat = np.zeros([total_time, total_users])
con = np.zeros([total_time, total_users])
# funcs = np.zeros([total_users, 2])

cpw[0] += 10

def f1(a):
    if a > 5: return 1
    if a < 0: return 0
    return 1/5*a

def f2(a):
    if a > 5: return 1
    if a < 0: return 0
    return 0.5

def f3(a):
    if a > 5: return 1
    if a < 0: return 0
    return -1/5*a+1

newATI = 2

t = 0
while t < total_time:
    for u in range(total_users):
        mat[t, u] = newATI
        if u == 0: contrib = f1(mat[t, u])
        if u == 1: contrib = f2(mat[t, u])
        if u == 2: contrib = f3(mat[t, u])
        if contrib < cpw[t, u]:
            con[t, u] = contrib
            cpw[t+1, u] = cpw[t, u] - contrib
    t += 1
    
print(np.sum(con))
```

```python
total_time = 100
total_users = 3
cpw = np.zeros([total_time+1, total_users])
mat = np.zeros([total_time, total_users])
con = np.zeros([total_time, total_users])
# funcs = np.zeros([total_users, 2])

cpw[0] += 10

def f1(a):
    if a > 5: return 1
    if a < 0: return 0
    return 1/5*a

def f2(a):
    if a > 5: return 1
    if a < 0: return 0
    return 0.5

def f3(a):
    if a > 5: return 1
    if a < 0: return 0
    return -1/5*a+1

t = 0
while t < total_time:
    for u in range(total_users):
        base = 4
        newATI = np.sum(con[:,u])/np.sum(con) if t != 0 else 1
        newATI = np.round(newATI * base) if t != 0 else 1
        if np.sum(mat[:,u]) > 100: 
            newATI = 0
        elif np.sum(mat[:,u]) + newATI > 100:
            newATI = 100 - np.sum(mat[:,u])            
        mat[t, u] = newATI
        if u == 0: contrib = f1(mat[t, u])
        if u == 1: contrib = f2(mat[t, u])
        if u == 2: contrib = f3(mat[t, u])
        if contrib < cpw[t, u]:
            con[t, u] = contrib
            cpw[t+1, u] = cpw[t, u] - contrib
    t += 1
    
print(np.sum(con))
```
